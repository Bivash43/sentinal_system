```python
import pandas as pd
df = pd.read_parquet('data/initial_normalized_data.parquet')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Let's look at the 'Class' column (0 = Legit, 1 = Fraud)
sns.countplot(x='Class', data=df)
plt.title('How many Frauds vs. Legit?')
plt.show()
```

```python
# We compare the 'Amount' for Fraud vs Legit
plt.figure(figsize=(10,6))
sns.boxplot(x='Class', y='normAmount', data=df)
plt.ylim(-1, 1) # Zooming in because some transactions are huge
plt.title('Do Fraudsters spend more?')
plt.show()
```

```python
plt.figure(figsize=(12,8))
# We check how much each 'V' column relates to 'Class'
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title('The Secret Pattern Map')
plt.show()
```

```python
# Let's see if fraud happens at certain times
plt.figure(figsize=(12,4))
sns.histplot(df[df['Class'] == 1]['normTime'], color='red', label='Fraud', kde=True)
sns.histplot(df[df['Class'] == 0]['normTime'], color='blue', label='Legit', kde=True, alpha=0.3)
plt.legend()
plt.show()
```
