```python
import pandas as pd

X_train = pd.read_parquet('data/X_train.parquet')
X_test = pd.read_parquet('data/X_test.parquet')
y_train = pd.read_parquet('data/y_train.parquet')
y_test = pd.read_parquet('data/y_test.parquet')
```

```python
from xgboost import XGBClassifier

# 1. Create the model
# We set 'n_estimators' to 100 (100 detectives in our team)
model = XGBClassifier(n_estimators=100, max_depth=4, learning_rate=0.1)

# 2. The "Study" phase (Training)
print("🧠 The AI is studying the patterns... please wait.")
model.fit(X_train, y_train)

print("✅ Training complete!")
```

```python
import joblib

# Save the model to a file
joblib.dump(model, 'model/sentinel_model.joblib')
print("💾 Model saved as 'sentinel_model.joblib'!")
```
