```python
import pandas as pd
df = pd.read_parquet('data/initial_normalized_data.parquet')
```

```python
# 1. Separate our 'Class 0' (Legit) and 'Class 1' (Fraud)
fraud_df = df[df['Class'] == 1]
legit_df = df[df['Class'] == 0]

# 2. We have 492 fraud cases.
# We will randomly grab 492 legit cases to match.
legit_sample = legit_df.sample(n=len(fraud_df), random_state=42)

# 3. Join them together into one new 'Balanced' dataset
balanced_df = pd.concat([fraud_df, legit_sample])

# 4. Shuffle them so they aren't all grouped together
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)
balanced_df.to_parquet('data/balanced_df.parquet')
# Let's see if it worked!
print("--- New Dataset Distribution ---")
print(balanced_df['Class'].value_counts())
```
